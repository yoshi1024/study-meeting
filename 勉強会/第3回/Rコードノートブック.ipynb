# 前処理

```r
df <- read.csv('./1212data/lesson1.csv',fileEncoding = 'sjis')
head(df)
```

```r
#注釈の列を削除
library(dplyr) 
df <- select(df, -注)
df
```

```r
#都道府県名がないデータは除外
df <- df[df$西暦.年.== 2015,]
df
```

```r
#欠損している行を削除
df <- na.omit(df)
df
```

```r
df <- df[
    (df$都道府県コード!='00') &
    (df$都道府県コード!='0A') &
    (df$都道府県コード!='0B'),
]

df
```

# グラフ作成

## 棒グラフ

```r
df$都道府県コード <- as.numeric(df$都道府県コード)
df
```

```r
df_touhoku <- df[df$都道府県コード < 8 ,]
df_touhoku
```

```r
# ggplot2パッケージ呼び出し
library(ggplot2)

# 作図
ggplot(data = df_touhoku, aes(x = 都道府県名, y = 人口.総数.)) + 
  geom_bar(stat = "identity") 
```
